```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.family"] = "Malgun Gothic"
mpl.rcParams["axes.unicode_minus"] = False
```

```python
csv = pd.read_csv("sea_rain.csv", encoding = "UTF8").set_index("연도")
csv
```

```python
csv.plot(grid = True)

plt.xticks(csv.index)
plt.xlabel("연도")
plt.ylabel("강수량")
plt.title("연간 강수량")

plt.show()
```
