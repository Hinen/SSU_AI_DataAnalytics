```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.family"] = "Malgun Gothic"
mpl.rcParams["axes.unicode_minus"] = False
```

```python
country = ["유럽", "미국", "캐나다", "호주", "중국"]
prefer = [7, 6, 3, 2, 2]

series = pd.Series(prefer, index = country, name = "")
series
```

```python
explodeList = (0.1, 0, 0, 0, 0)
pie = series.plot.pie(figsize = (5, 5), 
                            labels = country, 
                            autopct= '%.1f%%', 
                            startangle = 90,
                            counterclock = False, 
                            explode = explodeList, 
                            shadow = True)

pie.set_title("해외여행 선호도 조사 결과")
plt.show()
```
