```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import datasets
import matplotlib.pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
irs = datasets.load_iris()
irs # dic
irs.data # arr
```

```python
df = pd.DataFrame(irs.data)
df.head(10)
```

```python
data = pd.DataFrame({"length" : df[0], "width" : df[1]})
data.head(10)
```

```python
dataPoint = data.values
kmeans = KMeans(n_clusters = 3).fit(dataPoint)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
data['cluster_id'] = kmeans.labels_
data.head(10)
```

```python
df0 = data[data['cluster_id'] == 0]
df1 = data[data['cluster_id'] == 1]
df2 = data[data['cluster_id'] == 2]
```

```python
plt.scatter(df0['length'], df0['width'], c = "r", label = "0")
plt.scatter(df1['length'], df1['width'], c = "b", label = "1")
plt.scatter(df2['length'], df2['width'], c = "g", label = "2")

i = 0
for center in kmeans.cluster_centers_:
    color = "r"
    if i == 1:
        color = "b"
    elif i == 2:
        color = "g"
        
    plt.scatter(center[0], center[1], c = color, s = 300)
    i = i + 1

plt.xlabel("꽃받침 길이")
plt.ylabel("꽃받침 폭")
plt.legend()

plt.show()
```
