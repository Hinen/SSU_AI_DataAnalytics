```python
dataDic = { '정승현' : '바보', '김요셉' : '귀요미', '우수연' : '과탑' }
dataIndexList = ['호칭']
dataColumnsList = ['정승현', '김요셉', '우수연']

dataFrame = pd.DataFrame(dataDic, index = dataIndexList, columns = dataColumnsList)
dataFrame
```

```python
try:
    excelWriter = pd.ExcelWriter('data/excel_output.xlsx', engine='xlsxwriter')
    dataFrame.to_excel(excelWriter, sheet_name = '미경')
    excelWriter.save()
    
    print('엑셀 저장 완료')
except:
    print('혹시 엑셀을 열고있으면 꺼주세용')
```
