```python
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕으로 폰트 설정
matplotlib.rcParams['axes.unicode_minus'] = False 

df_rain = pd.read_csv('C:/myCode/data/sea_rain1.csv', index_col="연도" )
df_rain
```

```python
df_rain.plot()
plt.show()
```

```python
rain_plot = df_rain.plot(grid = True, style = ['r--*', 'g-o', 'b:*', 'm-.p'])
rain_plot.set_xticks([1996, 1997, 1998, 1999]) # x축 눈금
rain_plot.set_xlabel('연도') # x축 라벨
rain_plot.set_ylabel('강수량') # y축 라벨
rain_plot.set_title('연간강수량') #그래프 제목

plt.show()
```
