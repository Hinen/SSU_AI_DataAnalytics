```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/thoraric_surgery.csv")
df[0:10]
```

```python
dataArr = df.values
dataArr
```

```python
from sklearn.model_selection import train_test_split 

train, test = train_test_split(dataArr, test_size = 0.2)
xTrain = train[:, 0:17] # 학습 데이터
yTrain = train[:, 17] # 정답

xTest = train[:, 0:17] # 테스트용 데이터
yTest = train[:, 17] # 테스트용 정답
```

```python
xTrain
```

```python
yTrain[0:10]
```

```python
from keras.models import Sequential
from keras.layers import Dense

# 딥러닝 모델 Sequential
model = Sequential()

# 2개의 층 추가
# Dense(노드 수, input_dim = 데이터 갯수, activation = 활성화 함수 이름)
model.add(Dense(30, input_dim = 17, activation = "relu"))

# 출력층
model.add(Dense(1, activation = "sigmoid"))
```

```python
# 학습 환경 설정
# loss : 한번 신경망이 실행 될 때 마다 오차 값을 추적하는 함수
# mean_squared_error : 평균 제곱 오차
# optimizer : 오차를 어떻게 줄여 나갈지 정하는 함수 (가중치 설정 옵션)
# adam : 뭐지? 구글링 해봐야할듯 ㅇㅇ 강의에서 안 알려주네
# metrics : 학습과 테스트 단계에서 모델을 평가하기 위한 성능 기준 설정
# accuracy : 정확도를 기준으로 함
model.compile(loss = "mean_squared_error", 
              optimizer = "adam", 
              metrics = ["accuracy"])
```

```python
# 데이터를 이용해 학습 시작
# epochs : 몇번을 학습 할건지? -> 학습 횟수
# 학습 할수록 오차 loss가 줄어들고 정확도 accuracy가 증가함
# batch_size : 한번에 몇개의 데이터로 학습 할건지? 
model.fit(xTrain, yTrain, epochs = 30, batch_size = 10)
```

```python
# 모델이 얼마나 정확한지 예측 점검
# 해당 데이터에 대해 오차와 정확도가 출력됨
model.evaluate(xTest, yTest) # 데이터와 정답
```

```python
# 모델을 이용해 데이터를 이용해 예측 및 결과 출력
result = model.predict_classes(xTest)
result.ravel() # 무슨 다 죽어 ㅠㅠㅠㅋㅋㅋㅋㅋ 1이 없네
```

```python
resultDt = pd.DataFrame({
    "result" : result.ravel(), 
    "answer" : yTest.astype(int)
})

resultDt[0:10]
```
